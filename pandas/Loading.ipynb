```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
chest_pain = df.groupby(by="cp").median().reset_index()
chest_pain.head()
```

```python
chest_pain.plot.bar(x="cp", y="age")
```

```python
chest_pain.plot.bar(x="cp")
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label="age")
ax.set_xlabel("cp")
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.bar(chest_pain['cp'], chest_pain['age'], label="age", edgecolor="k", color=['red', 'blue', 'green', 'black'])
ax.set_xlabel("cp")
ax.legend()
```

```python
ax = sb.barplot('cp', 'age', errcolor="b", capsize=0.1, data=df);
```

# Scatter plots

```python
df.plot.scatter("age", "trestbps")
```

```python
fig, ax = plt.subplots()
ax.scatter(df['age'], df['trestbps'], marker="^", s=60, c=df['age'], edgecolors="k", alpha=0.5)
ax.set_xlabel('age')
ax.set_ylabel('trestbps')
```

```python
#sb.scatterplot('age', 'trestbps', hue="age", s=30, edgecolor='none', data=df)
sb.scatterplot(data=df, x='age', y='trestbps', hue='age', s=30, edgecolor="none")
```

```python
ages = df.groupby('age').median().reset_index()
ages.head()
```

```python
ages.plot.line('age', ['chol', 'trestbps'])
```

```python
fig, ax = plt.subplots()
ax.plot(ages['age'], ages['trestbps'], ls=":", lw=0.7)
ax.set_xlabel('age')
ax.set_ylabel('trestbps')
```

```python
fig, axes = plt.subplots(ncols=2)
df.plot.scatter(x='age', y='chol', ax=axes[0])
df.plot.scatter(x='age', y='trestbps', ax=axes[1])
fig.tight_layout()
```

```python
df
```

```python
fig2 = axes[0].get_figure()
fig2.savefig('plots.png', bbox_inches="tight", transparent=True)
```

```python
with plt.style.context('seaborn'):
    fig, axes = plt.subplots(ncols=3, sharey=True, gridspec_kw={'width_ratios': [2,1,1], 'wspace':0})
    y = 'age'
    xs = ['chol', 'trestbps', 'thalach']
    for x, ax in zip(xs, axes):
        ax.scatter(df[x], df[y])
        ax.set_xlabel(x)
    axes[0].set_ylabel(y)
```

```python
fig2 = axes[0].get_figure()
fig2.savefig('plots.png', bbox_inches="tight", transparent=True)
```
