# Modifying Type of Columns

```python
birthdate = pd.to_datetime(df['Birth Date'])
birthdate
```

```python
birthdate.dt.year
```

```python
zarya = pd.to_datetime('19981120')
df['age_at_zarya'] = (zarya - birthdate).astype('timedelta64[Y]')
df
```

```python
df['birth'] = birthdate
df.head(3)
```
