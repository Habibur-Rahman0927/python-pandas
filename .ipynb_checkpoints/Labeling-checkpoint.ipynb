# Rank

```python
dfp = df.sort_values('price', ascending=False)
dfp
```

```python
dfp['price_rank'] = df.price.rank(method="average", ascending=False)
dfp[['id', 'host_name', 'price', 'price_rank']].head(5)
```

```python
dfp['price_rank'] = df.price.rank(method="min", ascending=False)
dfp[['id', 'host_name', 'price', 'price_rank']].head(5)
```

```python
dfp['price_rank'] = df.price.rank(method="max", ascending=False)
dfp[['id', 'host_name', 'price', 'price_rank']].head(5)
```
